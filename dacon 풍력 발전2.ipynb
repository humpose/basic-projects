```python
#라이브러리 불러오기
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import os
import seaborn as sns
```

```python
#seed 고정
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(523)
```

```python
#데이터 불러오기
train = TabularDataset('/content/drive/MyDrive/풍력 발전_dacon/train.csv')
test = TabularDataset('/content/drive/MyDrive/풍력 발전_dacon/test.csv')
```

```python
train
```

```python
#train 데이터 확인
train.info()
```

```python
#train 데이터 기술통계량 확인
train.describe()
```

```python
train = train.drop(columns=['id', 'snowing'])
test=test.drop(columns=['id', 'snowing'])
```

```python
#모델 적합
predictor = TabularPredictor(label='target', eval_metric='mean_absolute_error').fit(train, presets='best_quality', time_limit=3600 * 5)
```

```python
# 모델 성능 확인
predictor.leaderboard(silent = True)
```

```python
#모델 예측
model_to_use = predictor.get_model_best()
model_pred = predictor.predict(test, model=model_to_use)
```

```python
#샘플파일 불러오기
sample=pd.read_csv("/content/drive/MyDrive/풍력 발전_dacon/sample_submission.csv")
```

```python
#예측값 입력
sample['target']=np.array(model_pred)
```
